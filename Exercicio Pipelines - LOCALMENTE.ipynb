```python
# FILTRANDO POR ATRASO NA **PARTIDA**COM TOTAL DE VEZES POR AEROPORTO DE ORIGEM
p1 = beam.Pipeline()

voos = (
    p1
    |'Extrair do CSV'>> beam.io.ReadFromText('/content/voos_filtro.csv',skip_header_lines = 1)
    |'Separador do CSV'>> beam.Map(lambda record: record.split(','))
    |'Filtrar cancelados'>> beam.Filter(lambda record: float(record[2]) == 0)
    |'Filtrar aero'>> beam.Filter(lambda record: float(record[1]) > 0)
    |'Agregar as Colunas'>> beam.Map(lambda record: (record[0],float(record[1])))
    |'Construir uma nova tablea por minuto'>> beam.CombinePerKey(sum)
    #|'Construir uma nova tablea por qtd'>> beam.combiners.Count.PerKey()
    |'Exibir o resultado'>> beam.Map(print)
    #|'Gravar o resultado'>> beam.io.WriteToText('saida.csv') #  NUNCA USAR MAP(PRINT) E WriteTo (ARQUIVO VAZIO)
)
p1.run()
```
